API call from Langchain or directly from provider

### Groq

- [List of models & rate limit](https://console.groq.com/docs/rate-limits)

```python
# pip install langchain-groq
import os
from langchain_groq import ChatGroq

GROQ_API_KEY = "..."
os.environ["GROQ_API_KEY"] = GROQ_API_KEY
```

```python
llm = ChatGroq(
    model="qwen/qwen3-32b",
    temperature=1.0,
    max_retries=2
)

llm_response = llm.invoke("hello")
```

### Google Gemini

- provider

```python
from google import genai

client = genai.Client(api_key="...")

response = client.models.generate_content(
    model="gemini-2.5-flash-lite",
    contents="hi"
)
print(response.text)
```

- Langchain

```python
# pip install langchain-google-genai
from langchain_google_genai import ChatGoogleGenerativeAI
model_api = "gemini-2.5-flash-lite"
GOOGLE_API_KEY = "..."
llm = ChatGoogleGenerativeAI(model=model_api, api_key=GOOGLE_API_KEY,
                             temperature=1.0,  # Gemini 3.0+ defaults to 1.0
                            max_tokens=None,
                            timeout=None,
                            max_retries=2)
```
