# Provider-Level API Calls

Direct LangChain calls to each provider — useful for testing individual connections.

**Providers:** [Groq](#groq) · [Google Gemini](#google-gemini) · [OpenRouter](#openrouter) · [Ollama Cloud](#ollama-cloud)

```python
from dotenv import load_dotenv
import os
load_dotenv(override=True)

# Or set keys directly:
# os.environ["GROQ_API_KEY"] = "..."
# os.environ["GOOGLE_API_KEY"] = "..."
# os.environ["OPENROUTER_API_KEY"] = "..."
# os.environ["OLLAMA_API_KEY"] = "..."
```

---
### Groq

- [Models & rate limits](https://console.groq.com/docs/rate-limits)

```python
from langchain_groq import ChatGroq

llm = ChatGroq(model="groq/compound-mini", temperature=1.0, max_retries=0)
response = llm.invoke("What date is today? What tool did you use to get this info?")
print(response.content)
```

---
### Google Gemini

- [Pricing & free tier](https://ai.google.dev/pricing)

**Native SDK**

```python
from google import genai

client = genai.Client()  # uses GOOGLE_API_KEY from env
response = client.models.generate_content(
    model="gemma-4-31b-it",
    contents="What date is today?"
)
print(response.text)
```

**LangChain**

```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite",
    temperature=1.0,
    max_retries=0
)
response = llm.invoke("Explain gravity in one sentence.")
print(response.content)
```

---
### OpenRouter

- [Free models](https://openrouter.ai/models?max_price=0)

```python
from langchain_openai import ChatOpenAI
from os import getenv

llm = ChatOpenAI(
    model="deepseek/deepseek-r1-0528:free",
    api_key=getenv("OPENROUTER_API_KEY"),
    base_url="https://openrouter.ai/api/v1",
    max_retries=0,
    default_headers={
        "HTTP-Referer": "https://github.com/tctsung/free-lunch-ai/",
        "X-Title": "free-lunch-ai"
    }
)
response = llm.invoke("Who are you?")
print(response.content)
```

---
### Ollama Cloud

Uses the OpenAI-compatible API at `https://ollama.com/v1/` — no extra dependency needed.

- [Cloud models](https://ollama.com/search?c=cloud)
- [API keys](https://ollama.com/settings/keys)
- [Pricing](https://ollama.com/pricing) (free tier: light usage, resets every 5 hours)

```python
from langchain_openai import ChatOpenAI
from os import getenv

llm = ChatOpenAI(
    model="gpt-oss:20b-cloud",
    api_key=getenv("OLLAMA_API_KEY"),
    base_url="https://ollama.com/v1/",
    max_retries=0
)
response = llm.invoke("Explain the no free lunch theorem in one sentence.")
print(response.content)
```
