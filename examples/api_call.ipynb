### Groq

- [List of models & rate limit](https://console.groq.com/docs/rate-limits)

```python
# pip install langchain-groq
# GROQ_API_KEY = "..."
# os.environ["GROQ_API_KEY"] = GROQ_API_KEY
```

```python
from langchain_groq import ChatGroq
llm = ChatGroq(
    model="groq/compound-mini",
    temperature=1.0,
    max_retries=2
)

llm_response = llm.invoke("what date is today? What tool did you use to get this info?")
print(llm_response.content)
```

### Google Gemini

- provider

```python
from google import genai

client = genai.Client()  # api_key="..."

response = client.models.generate_content(
    model="gemini-2.5-flash-lite",
    contents="what date is today"
)
print(response.text)
```

- Langchain

```python
# pip install langchain-google-genai
from langchain_google_genai import ChatGoogleGenerativeAI
model_api = "gemini-2.5-flash-lite"
llm = ChatGoogleGenerativeAI(model=model_api, 
                             temperature=1.0,  # Gemini 3.0+ defaults to 1.0
                            max_tokens=None,
                            timeout=None,
                            max_retries=2) # api_key=GOOGLE_API_KEY
```
